```python
import pandas as pd
import matplotlib.pyplot as plt
import sys
from pytz import timezone
import seaborn as sns
```

```python
df = pd.read_csv("data/cosimawellenbad-hallenbad.csv", header=None, names=["ts", "cnt", "max"], index_col="ts", parse_dates=["ts"])
df.index = df.index.tz_convert('Europe/Berlin')
```

```python
df['day_name'] = df.index.to_series().dt.weekday
df['hour'] = df.index.to_series().dt.hour
```

```python
pivot = pd.pivot_table(df[['day_name', 'hour', 'cnt']], index=['day_name', 'hour'], aggfunc='max')
```

```python
pivot = pivot.unstack(level=0)
```
